```python
# Seleziono le colonne di interesse per la migrazione
colonne_migrazione = [
    'Data',
    'ITALIA-Saldo migratorio interno (per mille)',
    'ITALIA-Saldo migratorio con l\'estero (per mille)',
    'ITALIA-Saldo migratorio totale (per mille)',
    'NORD-Saldo migratorio interno (per mille)',
    'NORD-Saldo migratorio con l\'estero (per mille)',
    'NORD-Saldo migratorio totale (per mille)',
    'CENTRO-Saldo migratorio interno (per mille)',
    'CENTRO-Saldo migratorio con l\'estero (per mille)',
    'CENTRO-Saldo migratorio totale (per mille)',
    'SUD-Saldo migratorio interno (per mille)',
    'SUD-Saldo migratorio con l\'estero (per mille)',
    'SUD-Saldo migratorio totale (per mille)',
    'ISOLE-Saldo migratorio interno (per mille)',
    'ISOLE-Saldo migratorio con l\'estero (per mille)',
    'ISOLE-Saldo migratorio totale (per mille)'
]

# Creo un DataFrame con le colonne di interesse per la migrazione
df_migrazione = df[colonne_migrazione]

# Calcolo la media della crescita della migrazione per l'Italia
media_saldo_migratorio_interno_italia = df_migrazione['ITALIA-Saldo migratorio interno (per mille)'].mean()
media_saldo_migratorio_estero_italia = df_migrazione['ITALIA-Saldo migratorio con l\'estero (per mille)'].mean()
media_saldo_migratorio_totale_italia = df_migrazione['ITALIA-Saldo migratorio totale (per mille)'].mean()

print(f"Media del saldo migratorio interno per l'Italia: {media_saldo_migratorio_interno_italia}")
print(f"Media del saldo migratorio con l'estero per l'Italia: {media_saldo_migratorio_estero_italia}")
print(f"Media del saldo migratorio totale per l'Italia: {media_saldo_migratorio_totale_italia}")

# Calcolo la media della crescita della migrazione per ciascuna regione
regioni = ['NORD', 'CENTRO', 'SUD', 'ISOLE']
for regione in regioni:
    media_saldo_migratorio_interno_regione = df_migrazione[f'{regione}-Saldo migratorio interno (per mille)'].mean()
    media_saldo_migratorio_estero_regione = df_migrazione[f'{regione}-Saldo migratorio con l\'estero (per mille)'].mean()
    media_saldo_migratorio_totale_regione = df_migrazione[f'{regione}-Saldo migratorio totale (per mille)'].mean()
    
    print(f"Media del saldo migratorio interno per la regione {regione}: {media_saldo_migratorio_interno_regione}")
    print(f"Media del saldo migratorio con l'estero per la regione {regione}: {media_saldo_migratorio_estero_regione}")
    print(f"Media del saldo migratorio totale per la regione {regione}: {media_saldo_migratorio_totale_regione}")
    
```
